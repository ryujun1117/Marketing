```python
def make_rgb(text):
    r = text.split(" ")[0].replace("R:","")
    g = text.split(" ")[1].replace("G:","")
    b = text.split(" ")[2].replace("B:","")
    return [int(r),int(g),int(b)]

def make_hsb(text):
    H = text.split(" ")[0].replace("H:","")
    S = text.split(" ")[1].replace("S:","")
    B = text.split(" ")[2].replace("B:","")
    return [int(H),int(S),int(B)]

def make_lab(text):
    l = text.split(" ")[0].replace("L*:","")
    a = text.split(" ")[1].replace("a*:","")
    b = text.split(" ")[2].replace("b*:","")
    return [float(l),float(a),float(b)]

def make_cmyk(text):
    c = text.split(" ")[0].replace("C:","").replace("%","")
    m = text.split(" ")[1].replace("M:","").replace("%","")
    y = text.split(" ")[2].replace("Y:","").replace("%","")
    k = text.split(" ")[3].replace("K:","").replace("%","")
    return [int(c),int(m),int(y),int(k)]

df2["rgb"] = df2["rgb"].apply(make_rgb)
df2["hsb"] = df2["hsb"].apply(make_hsb)
df2["lab"] = df2["lab"].apply(make_lab)
df2["cmyk"] = df2["cmyk"].apply(make_cmyk)
```
